```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
L_list = np.linspace(0,10,11)
C = 1
n = 1
x_0 = 1
r = 1 #r_tilde

x_prime = []
y_prime = []
for L in L_list:
    for theta in np.linspace(0,2*np.pi):
        delta_x = 2*L*C/n*x_0*r**2*np.sin(2*theta)
        delta_y = 2*L*C/n*x_0*r**2*(2+np.cos(2*theta))
        x_prime.append(delta_x)
        y_prime.append(delta_y)
print(delta_x)
plt.plot(x_prime, y_prime,'k')
   
```
